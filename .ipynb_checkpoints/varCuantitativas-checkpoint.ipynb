```python
# Creación aleatoria de datos
import numpy as np
X = np.array([0.5, 23, 0.3, 4.5, 0.3, 0.5,-28, -50, 60, -100, -10, -11, 13, 19, 1, 9])
print(X)
```

```python
# Medidas de tendencia central
from scipy import stats
media = np.mean(X)
mediana = np.median(X)
moda , _ = stats.mode(X)
print('media: ', media)
print('mediana: ', mediana)
print('moda :', moda)
```

```python
# Medidas de posición
import numpy as np
import matplotlib.pyplot as plt
Q1 = np.percentile(X, 25)
Q3 = np.percentile(X, 75)
RIC = Q3 - Q1
lim_inf = Q1 - 1.5*RIC
lim_sup = Q3 + 1.5*RIC
print('RIC :', RIC)
print('limite inferior :', lim_inf)
print('limite superior :', lim_sup)
```

```python
# Dibuja el diagrama de box and whiskers y determina visualmente el número de outliers
plt.boxplot(X)
plt.grid()
plt.title("Diagrama de caja y bigotes")
plt.show()
```

```python
# Medidas de dispersión
desvest = np.std(X)
print('Desviación típica: ', desvest)
print('Varianza: ', desvest**2)
```

```python
# Medidas de distribución
asimetria = stats.skew(X)
curtosis = stats.kurtosis(X, fisher=True) # https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kurtosis.html
print('Asimetría ', np.round(asimetria, 4))
print('Curtosis ', np.round(curtosis, 4))

if curtosis>0:
   print('Distribución Leptocúrtica')
elif curtosis<0:
   print('Distribución Platicúrtica')
else:
   print('Distribución Mesocúrtica')
```

```python
# EJERCICIO EXTRA SOBRE EL DATASET DE IRIS
# Extraer la media y la desviación típica de cada variable en función de la clase y representarlo en un gráfico (plt.errorbar)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets

# Cargamos el set de datos Iris
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df["target"] = iris.target

# Mapeamos los números de las clases a sus nombres reales
especies = iris.target_names  
df["especie"] = df["target"].map(lambda x: especies[x])

# Calculamos la media y desviación estándar agrupando por especie
medias = df.groupby("especie")[iris.feature_names].mean()
desviaciones = df.groupby("especie")[iris.feature_names].std()

plt.figure(figsize=(10, 6))

# Iterar sobre cada especie para graficar su línea de medias con barras de error
for especie in especies:
    plt.errorbar(
        x=iris.feature_names,
        y=medias.loc[especie],
        yerr=desviaciones.loc[especie],
        label=especie,
        fmt="-o",  
        capsize=5,  
        capthick=1.5,
    )

plt.title("Media y desviación estándar de cada variable por especie (Iris)")
plt.xlabel("Variables")
plt.ylabel("Valor (cm)")
plt.xticks(rotation=15) 
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend()
plt.tight_layout()

plt.show()
```
